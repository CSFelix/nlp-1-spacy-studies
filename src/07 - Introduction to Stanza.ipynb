Let's explore the basics of `Stanza` with a Language Model (LM) for `Wolof` language.

````python
# Downloading and Loading Language Model
import stanza

stanza.download(lang='wo', model_dir='./stanza-models') # wo: Wolof Language

nlp_wolof = stanza.Pipeline(lang='wo')
nlp_wolof_only_tokenizer_pos = stanza.Pipeline(lang='wo', processors='tokenize,pos')
````

````python
# Processing Single Document
document = nlp_wolof('R√©ew maa ngi lebe turam wi ci dex gi ko peek ci penku ak b√´j-g√†nnaar, te ab balluwaayam bawoo ca Fuuta Jallon ca Ginne, di Dexug Senegaal. Ab kilimaam bu g√´w√©el la te di bu fendi te yor √±aari jamono: jamonoy nawet (jamonoy taw) ak ju noor (jamonoy fendi).')
print(f'- Text: {document.text}')
print(f'- Tokens from Sentence 0: {document.sentences[0]}')
````

````python
document_dictionary = document.sentences[0].to_dict()
print(f'- Available Keys: {document_dictionary[0].keys}')
print(f'- Lemma: {document_dictionary[0]['lemma']}')
print(f'- All Attributes: {document_dictionary[0]}')
````

---

````python
# Processing Multiple Documents
texts = [
  'Lislaam a ngi nj√´kk a t√†bbi ci Senegaal ci diggante VIIIeelu xarnu ak IXeelu xarnu, √±a ko fa dugal di ay yaxantukat yu araab-yu-berber.'
  , 'Li √´pp ci g√´stu yi √±u def ci w√†llug G√´stu-askan (walla demogaraafi) ci Senegaal dafa sukkandiku ci Wa√±√± (recensement) yi √±u jotoon a def ci 1976, 1988 rawati na 2002.'
]

documents_wolof_input = [stanza.Document([], text=text) for text in texts]
print(f'- Document 0: {documents_wolof_input[0].text}')
print(f'- Document 1: {documents_wolof_input[1].text}')
````

````python
documents_wolof_output = nlp_wolof(documents_wolof_input)
print(f'- Processed Document 0 (with Annotations): {documents_wolof_output[0]}')
print(f'- Processed Document 1 (with Annotations): {documents_wolof_output[1]}')
````

<h1 id='1-spacy-stanza' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>‚òÅÔ∏è | Spacy-Stanza</h1>

`Spacy-Stanza` is a Python package that allows us to work with Spacy and Stanza NLP pre-trained models in a single interface!! In order to do it, we first must to install the package:

```bash
pip install -U spacy-stanza
```

Then we download a Language Model (LM) like we do in Stanza and we are ready to work with it.

There's only one observation: `the package only workds with languages that are available in both packages, Spacy and Stanza`. Besides, all generated objects follow Spacy's structure!!

Let's do an example with `Finnish` language, spoken in Finnish.

````python
# Spacy-Stanza
from spacy import displacy
import spacy
import stanza
import spacy_stanza

stanza.download(lang='fi', model_dir='./stanza-models')
nlp_fi = spacy_stanza.load_pipeline(name='fi')
````

````python
document = nlp_fi('Tove Jansson ker√§si 148 √§√§nt√§ eli 18,2% annetuista √§√§nist√§. Kirjailija, kuvataiteilija ja pilapiirt√§j√§ tuli kansainv√§lisesti tunnetuksi satukirjoistaan ja sarjakuvistaan.')
sentences = list(document.sents)
print(f'- Sentence 0: {sentences[0]}')
````

````python
displacy.render(document, style='dep')
````

````python
for token in sentences[0]:
    print(f'- Token: {token.text}')
    print(f'- Lemma: {token.lemma_}')
    print(f'- Coarse POS: {token.pos_}')
    print(f'- Fine-Grained POS: {token.tag_}')
    print(f'- Dependency Label: {token.dep_}')
    print(f'- Morpheme: {token.morph}')
    print('---')
````
